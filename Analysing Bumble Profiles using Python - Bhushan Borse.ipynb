### **Part 1 : Data Cleaning**

#### 1. Dealing with **Outliers** and **Invalid** datapoints in numerical columns

<small>Outliers are values that are much higher or lower than most of the other values in a dataset. They might be extreme due to natural variability (the result of natural differences in a population), Measurement, Data Entry or Experimental Errors, or unusual conditions.</small>

```python
# below code gives us the statistics of numerical columns present in provided dataset

bumble_df.describe().round(2)
```

<small>
<p>
    As it can be observed from above table <strong>age</strong> has a maximum value of <strong>110</strong> which is not possible in most cases. Hence this needs to be fixed. 
  </p>
  <p>
    There are several ways to detect outliers in numerical columns of a pandas DataFrame. Two popular methods are:
  </p>

  <ol>
    <li>Interquartile Range (IQR)</li>
    <li>Z‚Äëscore</li>
  </ol>

  <p><strong>Which Method to Choose?</strong></p>

  <p>
    <strong>IQR Method:</strong> This is non‚Äëparametric (it makes no assumptions about data distribution) and works well when data is skewed or when you have small to moderate sample sizes.
  </p>

  <p>
    <strong>Z‚ÄëScore Method:</strong> This method assumes that the data follows a roughly normal distribution and might be more appropriate if that assumption is met.
  </p>
</small>


<small>
<p>
  Detecting outliers in 'age' column using <strong>IQR method</strong>
  </p>
</small>

```python
# Calculate Q1, Q3 and the IQR for the column
Q1 = bumble_df['age'].quantile(0.25)
Q3 = bumble_df['age'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outlier rows
outliers = bumble_df[(bumble_df['age'] < lower_bound) | (bumble_df['age'] > upper_bound)]
    
# Store or print the outliers for the column
outliers_age_index = outliers.index.tolist()  # indices of outliers, if needed
print(f"Outliers in column '{'age'}':")
print(outliers[['age']])
```

<small>Using **Winsorization (Capping/ Flooring)** approach to "correct" outlier values in age column.

Winsorization (Capping/ Flooring): Replace any values below the lower bound with the lower bound value and those above the upper bound with the upper bound. This effectively "caps" the outliers.</small>

```python
# fixing outliers in age column

bumble_df.loc[bumble_df['age'] < lower_bound, 'age'] = lower_bound
bumble_df.loc[bumble_df['age'] > upper_bound, 'age'] = upper_bound
```

```python
bumble_df.describe()
```

<small>
<p>
Since <strong>height</strong> follows a roughly normal distribution we will be detecting outliers in height column using <strong>Z-Score method</strong>. For each value in height column we compute <strong>ùëß = (ùë•-ùúá)/ùúé</strong> and flag values with a ‚à£ùëß‚à£ greater than a threshold (commonly 3) as outliers.
  </p>
</small>

```python
import numpy as np

# Choose a threshold, commonly 3 (i.e. values with z-score > 3 or < -3 are considered outliers)
threshold = 3

mean_height = bumble_df['height'].mean()
std_height = bumble_df['height'].std()
    
# Calculate the z-scores for the column
z_scores = (bumble_df['height'] - mean_height) / std_height
print(z_scores)

# Identify outliers
outliers = bumble_df[np.abs(z_scores) > threshold]
    
print(f"\nOutliers in column '{'height'}' (using z-score):")

print(outliers[['height']])
```

<small>Using **Winsorization (Capping/ Flooring)** approach to "correct" outlier values in height column.

Since we're using a Z‚Äëscore threshold of 3, we can calculate the lower and upper bounds as follows:

Lower Threshold = mean_height ‚àí (3 √ó std_height)

Upper Threshold = mean_height + (3 √ó std_height)</small>

```python
# Calculate the lower and upper thresholds based on the z-score
lower_threshold = mean_height - (threshold * std_height)
upper_threshold = mean_height + (threshold * std_height)
```

```python
# fixing outliers in height column

bumble_df.loc[bumble_df['height'] < lower_threshold, 'height'] = lower_threshold
bumble_df.loc[bumble_df['height'] > upper_threshold, 'height'] = upper_threshold
```

```python
bumble_df.describe()
```

 <small>In the above summary, it can be seen that the 25th, 50th, and 75th percentiles for <strong>income</strong> column are all -1. This suggests that a very large fraction of records in provided dataset have an income value of -1. In many datasets, a value like -1 is used as a placeholder to indicate missing or <strong>‚Äúnot reported‚Äù</strong> income hence it‚Äôs a good idea to address these placeholders. </small>

<small>Here -1 represents missing data, hence we replace those with NaN so that statistical calculations focus on the reported (non-negative) incomes.</small>

```python
# Replacing -1 with NaN in the 'income' column
bumble_df['income'] = bumble_df['income'].replace(-1, np.nan)
```

```python
bumble_df['income'].describe()
```

<small>For income data, extreme high values (e.g., someone earning 1,000,000) can heavily skew the mean and other statistics. Hence we can use capping method to cap the extreme values to an upper threshold‚Äîfor example, the 95th percentile.</small>

```python
# Calculate the 95th percentile of income from the reported values
upper_cap = bumble_df['income'].quantile(0.95)
print(upper_cap)

# Cap incomes above this threshold to the 95th percentile value
bumble_df.loc[bumble_df['income'] > upper_cap, 'income'] = upper_cap
```

<small>Since 'age' could be a good indictor of educational qualification, we can utilise age to impute values in education column</small>

```python
# below shows count, min_age, mean_age and max_age for each education value in dataset

education_stats = bumble_df.groupby('education').agg(
    count=('education', 'count'),
    min_age=('age', 'min'),
    mean_age=('age', 'mean'),
    max_age=('age', 'max')
).reset_index()

education_stats = education_stats.sort_values(by='count', ascending=False)

education_stats
```

<small> Taking reference from above summary we can use below strategy to impute nulls in education column.

    if age>=36 then education == graduated from masters program	

    elif age between 29 to 36  then education == working on masters program	

    elif age between 25 to 29 then education == graduated from college/university	

    elif age <=25 then education == working on college/university</small>	

```python
def impute_education(age):
    if age >= 36:
        return "graduated from masters program"
    elif 29 <= age < 36:
        return "working on masters program"
    elif 25 <= age < 29:
        return "graduated from college/university"
    else:
        return "working on college/university"

# Imputing missing values:
bumble_df.loc[bumble_df['education'].isnull(), 'education'] = bumble_df.loc[bumble_df['education'].isnull(), 'age'].apply(impute_education)
```

#### 1. **Binning** and **Grouping**

<small>Grouping continuous variables, such as age or income, into bins helps simplify analysis and identify trends among specific groups. For instance, grouping users into age ranges can reveal distinct patterns in behavior or preferences across demographics. </small>

<small>Grouping ages into groups "18-25", "26-35", "36-45", "46+"</small>

```python
# Defining the bins and labels
bins = [18, 25, 35, 45, float('inf')]  

# Defining the category labels
labels = ["18-25", "26-35", "36-45", "46+"]  

# Create the 'age_group' column
bumble_df['age_group'] = pd.cut(bumble_df['age'], bins=bins, labels=labels, right=True)

# Display the first few rows to verify
print(bumble_df[['age', 'age_group']].head())
```

```python
sign_distribution = bumble_df.sign.value_counts()
sign_distribution
```

```python
import matplotlib.pyplot as plt
import squarify
import seaborn as sns

# Preparing data
sign_distribution_df = sign_distribution.reset_index()
sign_distribution_df = sign_distribution_df.sort_values(by='count', ascending=True)
sign_distribution_df.columns = ['sign', 'count']  # Rename columns for clarity

# Create a custom color palette with normalized values
color_palette = sns.color_palette('pastel', len(sign_distribution_df))  # Generate diverse colors (in range [0, 1])

# Create the treemap
plt.figure(figsize=(25, 19))
squarify.plot(
    sizes=sign_distribution_df['count'],
    label=[f"{sign}\n({count})" for sign, count in zip(sign_distribution_df['sign'], sign_distribution_df['count'])],
    color=color_palette,  # Use the normalized color palette
    alpha=0.8,
    text_kwargs={'fontsize': 10}  # Reduce label font size
)

# Add title
plt.title('Sign Distribution Treemap', fontsize=14)

# Remove axes for a cleaner look
plt.axis('off')

# Display the plot
plt.show()
```
